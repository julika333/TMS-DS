### Что такое рекомендательные системы и зачем они нужны?

Рекомендательные системы - это класс алгоритмов и методов, используемых для предоставления пользователю рекомендаций о том, какие товары, услуги или контент могут его заинтересовать. Они играют ключевую роль в современных онлайн-сервисах, позволяя улучшить опыт пользователей и повысить эффективность бизнеса.

**Зачем они нужны?**

1. **Пользовательский опыт**: Рекомендательные системы позволяют предоставлять пользователям более персонализированный контент. Это способствует повышению удовлетворенности пользователей и удержанию их на платформе.

2. **Увеличение продаж**: В бизнесе рекомендации могут значительно увеличить продажи. Когда система предлагает пользователю товары, которые могут его заинтересовать, вероятность покупки значительно возрастает.

3. **Долгосрочные отношения с клиентами**: Рекомендательные системы помогают создать более тесные связи с клиентами. Пользователи чувствуют, что сервис заботится о них, предоставляя релевантные рекомендации.

4. **Фильтрация информации**: В мире с избытком информации, рекомендации помогают пользователям найти наиболее интересующий их контент, минимизируя необходимость поиска.

5. **Увеличение вовлеченности**: Рекомендательные системы могут помочь удерживать пользователей на платформе, предлагая им интересный и актуальный контент.

**Примеры использования:**

- **Интернет-магазины**: Рекомендации товаров и услуг, основанные на предыдущих покупках и просмотрах.
- **Стриминговые сервисы**: Рекомендации фильмов, сериалов и музыки, основанные на предпочтениях пользователя.
- **Социальные сети**: Рекомендации друзей, групп и контента, которые могут заинтересовать пользователя.
- **Платформы новостей**: Рекомендации новостей и статей в соответствии с интересами пользователя.

# Принципы рекомендаций на основе пользователей

Рекомендации на основе пользователей — это метод, при котором предложения делаются на основе похожих интересов между пользователями.

## 1. Как это работает?

### a. Расчет похожести

Для каждого пользователя системы рассчитывается мера похожести с другими пользователями.

### b. Прогнозирование

На основе оценок похожих пользователей прогнозируется, какие продукты или услуги будут интересны данному пользователю.

## 2. Методы расчета похожести

* **Евклидово расстояние**: Чем меньше расстояние, тем больше похожесть.
* **Коэффициент корреляции Пирсона**: Измеряет линейную зависимость между двумя наборами данных.

## 3. Преимущества

* Простота реализации.
* Персонализированные рекомендации.

## 4. Недостатки

* Проблема "холодного старта": новые пользователи без истории взаимодействия.
* Высокая вычислительная сложность при большом количестве пользователей.

## 5. Примеры применения

* Рекомендация книг или фильмов.
* Музыкальные сервисы, такие как Spotify или Apple Music.

## Заключение

Рекомендации на основе пользователей ориентированы на поиск схожих пользователей для предоставления релевантных предложений. Они эффективны в система


# Математические основы оценки сходства пользователей

Оценка сходства пользователей является ключевым компонентом многих систем рекомендаций. Давайте рассмотрим некоторые основные методы.

## 1. Евклидово расстояние

Используется для измерения "расстояния" между двумя точками в пространстве.

### Формула:
$ d(p, q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2 + ... + (p_n - q_n)^2}$

где $p$ и $q$ — векторы оценок двух пользователей, а $n$ — количество элементов в векторе.

## 2. Коэффициент корреляции Пирсона

Измеряет линейную зависимость между двумя наборами данных.

### Формула:
\$ r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \sum{(Y_i - \bar{Y})^2}}} \$

где $X_i$ и $Y_i$ — оценки двух пользователей, а $\bar{X}$ и $\bar{Y}$ — средние значения этих оценок.

## 3. Косинусное сходство

Измеряет косинус угла между двумя векторами. Используется чаще всего в текстовых системах рекомендаций.

### Формула:
\$\text{similarity} = \frac{A \cdot B}{\|A\| \|B\|} \$
где $A$ и $B$ — векторы оценок двух пользователей.

## 4. Манхэттенское расстояние (расстояние городских кварталов)

Используется для измерения сходства на основе суммы модулей разностей координат.

### Формула:
\$ d(p, q) = \sum_{i=1}^{n} |p_i - q_i| \$

Выбор метода оценки сходства зависит от специфики задачи и типа данных. Важно правильно интерпретировать результаты, чтобы обеспечить релевантные рекомендации.

````python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Загрузка данных
movies_metadata = pd.read_csv('C:/Users/Sidromnik/DS_course/Untitled Folder 1/data/movies_metadata.csv', low_memory=False)
movies = movies_metadata[['title', 'overview']].dropna().reset_index(drop=True)

# Векторизация описаний
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movies['overview'])

# Расчет косинусного сходства
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movies.index[movies['title'] == title].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Получаем 10 самых похожих фильмов
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]

# Пример использования:
print(get_recommendations('The Dark Knight'))
````

# Принципы рекомендаций на основе контента

Рекомендации на основе контента — это подход, при котором предложения делаются на основе анализа содержания или атрибутов элементов.

## 1. Основная идея

Рекомендации на основе контента опираются на схожесть между содержимым элементов (например, фильмов, книг, товаров) и интересами или предпочтениями пользователя.

## 2. Как это работает?

### a. Анализ контента

Сначала проводится анализ контента каждого элемента. Это может включать в себя:

- Текстовый анализ (например, ключевые слова из описания продукта).
- Анализ изображений или видео.
- Анализ атрибутов товара (например, жанры фильма).

### b. Профиль пользователя

На основе взаимодействий пользователя (например, оценок, покупок или просмотров) создается профиль пользователя, который содержит информацию о его интересах.

### c. Совпадение

Сравниваются профили пользователей и контент элементов, чтобы найти наиболее подходящие элементы для рекомендации.

## 3. Преимущества

- **Персонализация**: рекомендации на основе уникальных интересов пользователя.
- **Независимость от других пользователей**: рекомендации не зависят от оценок или предпочтений других пользователей.

## 4. Недостатки

- **Ограниченная разнообразность**: рекомендации могут ограничиваться только тем, что пользователь уже видел или оценивал.
- **Проблема "холодного старта"**: сложно рекомендовать что-то новым пользователям без истории взаимодействия.

## 5. Примеры применения

- Рекомендации статей на новостных сайтах.
- Рекомендация продукции на онлайн-платформах в соответствии с интересами пользователя.

# Математические основы определения сходства объектов (20 мин)

Определение сходства между объектами является ключевой частью многих систем рекомендаций и алгоритмов машинного обучения. Есть разные методы измерения сходства.

## 1. Евклидово расстояние

Евклидово расстояние между двумя точками в евклидовом пространстве определяется следующим образом:

$ d(\mathbf{p}, \mathbf{q}) = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2} $

где:
- $ \mathbf{p} $ и $ \mathbf{q} $ — векторы в n-мерном пространстве.

Меньшее значение Евклидового расстояния указывает на большее сходство между объектами.

## 2. Косинусное сходство

Косинусное сходство измеряет косинус угла между двумя векторами. Оно определяется следующим образом:

$ \text{sim}(\mathbf{A}, \mathbf{B}) = \frac{\mathbf{A} \cdot \mathbf{B}}{||\mathbf{A}|| \times ||\mathbf{B}||} $

где:
- $ \mathbf{A} $ и $ \mathbf{B} $ — векторы.
- $ \mathbf{A} \cdot \mathbf{B} $ — скалярное произведение векторов.
- $ ||\mathbf{A}|| $ — евклидова норма (или длина) вектора $ \mathbf{A} $.

Косинусное сходство находится в диапазоне от -1 до 1, где 1 указывает на полное сходство, а -1 — на полное различие.

## 3. Манхэттенское расстояние

Манхэттенское расстояние между двумя точками в n-мерном пространстве определяется как:

$ d(\mathbf{p}, \mathbf{q}) = \sum_{i=1}^{n} |q_i - p_i| $

Это расстояние представляет собой сумму абсолютных различий между координатами точек.

## 4. Расстояние Жаккара

Расстояние Жаккара определяется для множеств:

$ J(A, B) = 1 - \frac{|A \cap B|}{|A \cup B|} $

где $ A $ и $ B $ — множества. Значение 0 указывает на полное сходство множеств, а 1 — на полное их различие.

Определение сходства между объектами является фундаментальным для многих доменов, включая рекомендательные системы, классификацию и кластеризацию. Выбор метода измерения сходства зависит от типа данных и конкретной задачи.

# Часть 3: Коллаборативный фильтр
Введение в коллаборативный фильтринг
Коллаборативный фильтринг — это метод, используемый в рекомендательных системах, где рекомендации делаются на основе предыдущего поведения пользователей. В основном, это метод находит схожих пользователей или элементов, чтобы дать рекомендации.

## Основные идеи:

- User-based (на основе пользователей): Если пользователь $A$ похож на пользователя $B$, и пользователь $A$ понравился определенный продукт, который пользователь $B$ еще не видел, то продукт может быть рекомендован пользователю $B$.

- Item-based (на основе элементов): Если пользователи, которые понравились элементы $X$ и $Y$, также понравились элемент $Z$, то у пользователя, которому понравились элементы $X$ и $Y$, может быть рекомендован элемент $Z$.

- Преимущества:

Не требуется никакой информации о пользователях или элементах (кроме истории взаимодействия).
Может дать очень точные рекомендации для пользователей, у которых есть достаточно истории взаимодействия.
Недостатки:

Проблема "холодного старта": сложно давать рекомендации для новых пользователей или новых элементов.
Масштабируемость: вычислительные сложности могут быстро расти с увеличением числа пользователей или элементов.
В последующих разделах мы рассмотрим, как реализовать основные методы коллаборативного фильтра с использованием различных инструментов и подходов.

# User-Based и Item-Based Подходы в Рекомендательных Системах

Рекомендательные системы стали важной частью многих онлайн-платформ. Они помогают пользователям находить интересующие их продукты или контент, а также увеличивают продажи для бизнесов. Основные подходы к созданию рекомендательных систем - это user-based (основанный на пользователях) и item-based (основанный на элементах) подходы.

## User-Based Подход

User-based подход, как следует из названия, фокусируется на схожести между пользователями.

### Принцип:

1. Вычислить схожесть между пользователями на основе их оценок или поведения.
2. Найти наиболее похожих пользователей для данного пользователя.
3. Рекомендовать элементы, которые нравятся этим похожим пользователям.

### Математическая формула схожести:

Самый популярный метод вычисления схожести - коэффициент корреляции Пирсона:

$ \text{sim}(u, v) = \frac{\sum_{i \in I}(r_{ui} - \bar{r_u})(r_{vi} - \bar{r_v})}{\sqrt{\sum_{i \in I}(r_{ui} - \bar{r_u})^2 \sum_{i \in I}(r_{vi} - \bar{r_v})^2}} $

где:
- $ \text{sim}(u, v) $ - схожесть между пользователем $u$ и пользователем $v$.
- $ r_{ui} $ - оценка, данная пользователем $u$ элементу $i$.
- $ \bar{r_u} $ - среднее значение оценок пользователя $u$.

## Item-Based Подход

В отличие от user-based, item-based подход фокусируется на схожести между элементами, а не пользователями.

### Принцип:

1. Вычислить схожесть между элементами на основе оценок пользователей.
2. Для данного элемента найти наиболее похожие элементы.
3. Рекомендовать эти элементы пользователям.

### Математическая формула схожести:

Также можно использовать коэффициент корреляции Пирсона, но с учетом элементов:

$ \text{sim}(i, j) = \frac{\sum_{u \in U}(r_{ui} - \bar{r_i})(r_{uj} - \bar{r_j})}{\sqrt{\sum_{u \in U}(r_{ui} - \bar{r_i})^2 \sum_{u \in U}(r_{uj} - \bar{r_j})^2}} $

где:
- $ \text{sim}(i, j) $ - схожесть между элементом $i$ и элементом $j$.
- $ r_{ui} $ - оценка, данная пользователем $u$ элементу $i$.
- $ \bar{r_i} $ - среднее значение оценок для элемента $i$.

В заключении, выбор между этими подходами зависит от конкретных задач и доступных данных. Иногда комбинированный подход может дать наилучшие результаты.

# Матричное разложение в рекомендательных системах: роль SVD

Матричное разложение стало ключевым инструментом в рекомендательных системах из-за его способности выделять скрытые факторы, которые определяют предпочтения пользователей.

## Основы матричного разложения

Матричное разложение, в своей основе, представляет собой разбиение одной матрицы на произведение нескольких матриц. В контексте рекомендательных систем, исходная матрица обычно представляет собой матрицу пользователь-предмет, где каждый элемент отражает рейтинг или предпочтение конкретного пользователя к конкретному предмету.

## SVD (Singular Value Decomposition)

SVD является одним из наиболее популярных методов матричного разложения, применяемых в рекомендательных системах. Математически SVD разбивает матрицу $A$ на три другие матрицы:

$ A = U \Sigma V^T $

где:
- $U$ — матрица левых сингулярных векторов (относится к пользователям).
- $\Sigma$ — диагональная матрица с сингулярными значениями.
- $V^T$ — матрица правых сингулярных векторов (относится к элементам).

## Преимущества использования SVD в рекомендательных системах

1. **Выявление скрытых факторов**: С помощью SVD можно выявить скрытые факторы или концепции, которые могут объяснить предпочтения пользователей. Например, в случае фильмов эти факторы могут быть связаны с жанрами, актерами или режиссерами.

2. **Снижение размерности**: Используя только первые $k$ сингулярных значений (и соответствующие векторы), можно приближенно восстановить исходную матрицу. Это позволяет уменьшить объем данных без существенной потери информации.

3. **Прогнозирование отсутствующих значений**: SVD позволяет прогнозировать значения в матрице пользователь-предмет, которые изначально отсутствовали, что полезно для предложения рекомендаций пользователям.

SVD предоставляет эффективный способ обработки больших матриц оценок и выявления взаимосвязей между пользователями и элементами. Благодаря этим качествам SVD стал важным инструментом в современных рекомендательных системах.


## Теоретические основы SVD
- Введение в SVD (Singular Value Decomposition).
- Математическая формулировка:
  $ A = U \Sigma V^T $
  где:
  - $A$ — исходная матрица оценок пользователей.
  - $U$ — левая сингулярная матрица (связана с пользователями).
  - $\Sigma$ — диагональная матрица с сингулярными значениями.
  - $V^T$ — правая сингулярная матрица (связана с элементами).

- Значение сингулярных значений и их отношение к "энергии" или важности информации.

## Применение SVD в рекомендательных системах
- Проблемы прямого применения SVD к матрицам рейтингов (например, пропущенные значения).
- Использование SVD для прогнозирования оценок:
  $ \hat{A} = U_k \Sigma_k V_k^T $
  где $k$ — число сингулярных значений, используемых для приближения.

# Пример
Использование SVD на The Movies Dataset включает в себя несколько этапов: загрузка и предварительная обработка данных, создание матрицы пользователь-предмет, применение SVD и, наконец, генерацию рекомендаций.

Приведу пример кода на Python с использованием библиотеки pandas и scipy:

````python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import svds

# Загрузка данных
movies = pd.read_csv('C:/Users/Sidromnik/DS_course/Untitled Folder 1/data/movies_metadata.csv', low_memory=False)
ratings = pd.read_csv('C:/Users/Sidromnik/DS_course/Untitled Folder 1/data/ratings_small.csv')
````

````python
# Преобразование столбца 'id' в числовой формат
movies['id'] = pd.to_numeric(movies['id'], errors='coerce')

# Создание матрицы пользователь-предмет
user_ratings = ratings.pivot(index='userId', columns='movieId', values='rating')
user_ratings = user_ratings.fillna(0)

# Центрирование рейтингов
user_ratings_mean = user_ratings.mean(axis=1)
user_ratings_demeaned = user_ratings - user_ratings_mean.values.reshape(-1, 1)

# Применение SVD
U, sigma, Vt = svds(user_ratings_demeaned.values, k=50)

# Восстановление прогнозируемой матрицы
sigma = np.diag(sigma)
predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.values.reshape(-1, 1)
predicted_ratings_df = pd.DataFrame(predicted_ratings, columns=user_ratings.columns)

# Функция для рекомендаций
def recommend_movies(user_id, num_recommendations=5):
    user_row = user_id - 1  # userId начинается с 1, не с 0
    sorted_user_predictions = predicted_ratings_df.iloc[user_row].sort_values(ascending=False)

    # Получить данные пользователя и объединить с данными о фильмах
    user_data = ratings[ratings.userId == user_id]
    user_full = (user_data.merge(movies, how='left', left_on='movieId', right_on='id')
                 .sort_values(['rating'], ascending=False))

    # Рекомендации для пользователя
    recommendations = (movies[~movies['id'].isin(user_full['movieId'])]
                       .merge(pd.DataFrame(sorted_user_predictions).reset_index(), how='left', left_on='id', right_on='movieId')
                       .rename(columns={user_row: 'Predictions'})
                       .sort_values('Predictions', ascending=False)
                       .iloc[:num_recommendations, :-1])

    return user_full, recommendations

# Тестирование функции
user_full, recommendations = recommend_movies(1, 5)
print(recommendations)
````

- **RMSE (Root Mean Squared Error)**: $RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$
- **Precision@k**: $Precision@k = \frac{\text{Number of relevant items in top k}}{\text{k}}$
- **Recall@k**: $Recall@k = \frac{\text{Number of relevant items in top k}}{\text{Total number of relevant items}}$
- **Hit Rate**: Определение может варьироваться, но часто это: $Hit \ Rate = \frac{\text{Number of hits}}{\text{Total number of recommendations}}$
- **Diversity**: Определение также может варьироваться, но часто это измеряется как среднее разнообразие между всеми парами элементов в рекомендациях.


````python
from sklearn.metrics import mean_squared_error
from math import sqrt

def rmse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten()
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, ground_truth))
````

````python
def precision_recall_at_k(predictions, k=10, threshold=3.5):
    user_est_true = defaultdict(list)
    for uid, _, true_r, est, _ in predictions:
        user_est_true[uid].append((est, true_r))

    precisions = dict()
    recalls = dict()
    for uid, user_ratings in user_est_true.items():
        user_ratings.sort(key=lambda x: x[0], reverse=True)
        n_rel = sum((true_r >= threshold) for (_, true_r) in user_ratings)
        n_rec_k = sum((est >= threshold) for (est, _) in user_ratings[:k])
        n_rel_and_rec_k = sum(((true_r >= threshold) and (est >= threshold)) for (est, true_r) in user_ratings[:k])

        precisions[uid] = n_rel_and_rec_k / n_rec_k if n_rec_k != 0 else 1
        recalls[uid] = n_rel_and_rec_k / n_rel if n_rel != 0 else 1

    return precisions, recalls
````

````python
def hit_rate(predictions, threshold=3.5):
    hits = 0
    total = 0

    for uid, _, true_r, est, _ in predictions:
        if true_r >= threshold:
            total += 1
            if est >= threshold:
                hits += 1

    return hits / total
````

````python
def diversity(predictions, num_items):
    rec_items_set = set([x[1] for x in predictions])
    overlap = 0
    for item in rec_items_set:
        for other_item in rec_items_set:
            if item != other_item:
                overlap += int(item == other_item)
    diversity = 1 - (overlap / (len(rec_items_set) * (len(rec_items_set) - 1)))
    return diversity
````

Эти функции предполагают наличие predictions, которые являются списком кортежей (uid, iid, true_r, est, details), где:

uid: идентификатор пользователя
iid: идентификатор элемента (в нашем случае фильма)
true_r: истинный рейтинг
est: оцененный рейтинг
details: детали оценки (не всегда требуются)
Эти функции могут быть адаптированы для конкретных потребностей или данных.
