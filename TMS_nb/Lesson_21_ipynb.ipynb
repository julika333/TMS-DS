# Математические основы
Мера схожести (или расстояния)

Для того чтобы кластеризовать данные, нам нужно определить, как измерять "похожесть" или "расстояние" между объектами.

## Евклидово расстояние между двумя точками $p$ и $q$ в $n$-мерном пространстве:
$d(p,q) = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2}$

## Манхэттенское расстояние (или L1 норма):
$d(p,q) = \sum_{i=1}^{n} |q_i - p_i|$

## Косинусное сходство между двумя векторами $A$ и $B$:
$\text{similarity}(A,B) = \frac{A \cdot B}{|A|_2 |B|_2}$
где $A \cdot B$ представляет скалярное произведение, а $|A|_2$ - это L2 норма вектора $A$.

# Объекты и признаки

Рассмотрим набор данных, состоящий из $m$ объектов, каждый из которых имеет $n$ признаков. Тогда этот набор данных можно представить в виде матрицы размером $m \times n$, где каждая строка представляет объект, а каждая колонка представляет признак.

# Функционал качества кластеризации

Для многих алгоритмов кластеризации целью является минимизация или максимизация некоторого функционала качества. Например, для метода k-средних цель заключается в минимизации суммарного квадратичного отклонения точек от центров их кластеров:
$J = \sum_{i=1}^{k} \sum_{x \in C_i} | x - \mu_i |^2$

где:

$k$ — число кластеров,
$C_i$ — i-й кластер,
$\mu_i$ — центр i-го кластера.

# Метод k-средних

Теоретическое введение

Определение и основная идея:

Метод k-средних — это итерационный алгоритм, цель которого заключается в разделении набора данных на $k$ кластеров, где каждый кластер представлен своим центром (или "средним").

Мера расстояния:

Наиболее часто используемая мера расстояния в методе k-средних — это евклидово расстояние:
$d(p, q) = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2}$

Алгоритмический процесс:

Инициализация: Выбор $k$ начальных центров кластеров случайным образом или с использованием какого-либо умного метода (например, k-means++).
Итерация:
Присвоение каждой точке к ближайшему центру кластера.
Пересчет центров кластеров как средних значений всех точек, принадлежащих этому кластеру.
Остановка: Когда центры кластеров перестают существенно изменяться или после заданного числа итераций.
Функционал качества:

Функционал качества для метода k-средних обычно определяется как суммарное квадратичное отклонение точек от центров их кластеров:
$J = \sum_{i=1}^{k} \sum_{x \in C_i} | x - \mu_i |^2$
где:

$J$ — функционал качества,
$C_i$ — i-й кластер,
$\mu_i$ — центр i-го кластера.
Проблемы и ограничения:

Выбор подходящего числа $k$ может быть неявным.
Чувствительность к выбору начальных центров.
Может не справляться с кластерами несферической формы.
Заключение: Метод k-средних — это популярный и легко понимаемый метод кластеризации, который может быть очень эффективным на больших наборах данных. Однако, важно помнить о его ограничениях и выбирать настройки метода в зависимости от конкретной задачи.

# Метрики качества кластеризации
Оценка качества кластеризации — важный этап в процессе анализа данных. Используется несколько ключевых метрик, которые помогают определить, насколько хорошо алгоритм справился с задачей:

- Силуэтный коэффициент (Silhouette Coefficient)
Силуэтный коэффициент измеряет, насколько объект похож на свой собственный кластер по сравнению с другими кластерами. Значения коэффициента лежат в диапазоне от -1 до 1, где высокое значение указывает, что объект хорошо сопоставлен с собственным кластером и плохо сопоставлен с соседними кластерами.

Формула:

$ s(i) = \frac{b(i) - a(i)}{\max{a(i), b(i)}}$

где:

$a(i)$ — среднее расстояние от i-го объекта до других объектов в том же кластере.
$b(i)$ — наименьшее среднее расстояние от i-го объекта до объектов в другом кластере, отличном от того, в котором находится i.
- Индекс Данна (Davies-Bouldin Index)
Этот индекс измеряет среднее "расстояние" между кластерами, где меньшие значения соответствуют лучшей кластеризации.

Формула:

$ DB = \frac{1}{k} \sum_{i=1}^{k} \max_{i \neq j} \left( \frac{\sigma_i + \sigma_j}{d(c_i, c_j)} \right)$

где:

$k$ — число кластеров.
$\sigma_i$ — среднее расстояние между каждым элементом кластера i и центроидом этого кластера.
$d(c_i, c_j)$ — расстояние между центроидами кластеров i и j.
- Оценка качества кластеризации на основе плотности (DBSCAN)
В случае использования алгоритмов, основанных на плотности, таких как DBSCAN, качество часто оценивается на основе числа образованных кластеров, числа выбросов и плотности кластеров.

Однако учитывайте, что многие метрики качества кластеризации зависят от конкретной природы данных и задачи. Не существует "универсальной" метрики, которая была бы наилучшей для всех задач кластеризации.


### Определение оптимального значения $k$
(числа кластеров) — одна из основных проблем в алгоритмах кластеризации, особенно в методе k-средних. Вот несколько популярных методов определения оптимального значения $k$:

- Метод "локтя" (Elbow method)
Метод заключается в запуске кластеризации на наборе данных для различного числа кластеров (обычно от 1 до n) и измерении суммы квадратов расстояний от каждой точки до её центра. Когда этот показатель резко замедляет свой рост, это может указывать на то, что добавление дополнительных кластеров не дает значительных преимуществ.

Формула для метода k-средних:

$ WCSS = \sum_{i=1}^{k} \sum_{x \in C_i} | x - \mu_i |^2$

где:

$WCSS$ — Within-Cluster-Sum-of-Squares.
$C_i$ — i-й кластер.
$\mu_i$ — центроид i-го кластера.
Метод силуэтного коэффициента (Silhouette method)
Этот метод использует среднее значение силуэтного коэффициента для всех точек. Если $k$ выбрано правильно, то этот показатель должен быть максимален.

- Формула силуэтного коэффициента:

$ s(i) = \frac{b(i) - a(i)}{\max{a(i), b(i)}}$

где:

$a(i)$ — среднее расстояние от i-го объекта до других объектов в том же кластере.
$b(i)$ — наименьшее среднее расстояние от i-го объекта до объектов в другом кластере, отличном от того, в котором находится i.
- Метод "разреженной матрицы" (Gap statistic)
Этот метод сравнивает изменение в общем распределении входных данных с изменением для случайного образца. Оптимальное число кластеров соответствует минимальной разнице между этими распределениями.

В зависимости от природы данных и задачи, один метод может работать лучше другого. По этой причине рекомендуется использовать несколько методов в комбинации, чтобы определить наиболее подходящее значение $k$ для вашего конкретного случая.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Генерация простых данных для демонстрации
data, _ = make_blobs(n_samples=3000, centers=6, cluster_std=0.60, random_state=0)

# Список для хранения WCSS (сумма квадратов расстояний внутри кластера)
wcss = []

# Применим k-средних для k от 1 до 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

# Визуализация результатов
plt.plot(range(1, 11), wcss)
plt.title('Метод "Локтя"')
plt.xlabel('Число кластеров')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Генерация простых данных для демонстрации
data, _ = make_blobs(n_samples=500, centers=3, cluster_std=0.60, random_state=0)

# Список для хранения силуэтных коэффициентов
silhouette_coefficients = []

# Начинаем с минимума 2 кластеров, так как для одного кластера силуэтный коэффициент не определен
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    score = silhouette_score(data, kmeans.labels_)
    silhouette_coefficients.append(score)

# Визуализация результатов
plt.plot(range(2, 11), silhouette_coefficients)
plt.title('Метод силуэтного коэффициента')
plt.xlabel('Число кластеров')
plt.ylabel('Силуэтный коэффициент')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt

def gap_statistic(data, max_clusters=10, n_references=5):
    # Создание случайных образцов данных
    shape = data.shape
    tops = data.max(axis=0)
    bots = data.min(axis=0)
    dists = np.diag(tops - bots)

    # Оценка на исходных данных
    gaps = np.zeros((len(range(1, max_clusters)),))
    resultsdf = np.zeros((len(range(1, max_clusters)),))
    for (i, k) in enumerate(range(1, max_clusters)):
        km = KMeans(n_clusters=k)
        km.fit(data)

        disp = sum(np.min(cdist(data, km.cluster_centers_, 'euclidean'), axis=1)) / shape[0]
        resultsdf[i] = disp

        # Оценка на случайных данных
        refdisps = np.zeros(n_references)
        for j in range(n_references):
            rdata = (np.random.random_sample(size=shape) * dists[0, :]) + bots
            km.fit(rdata)
            refdisps[j] = sum(np.min(cdist(rdata, km.cluster_centers_, 'euclidean'), axis=1)) / shape[0]
        gaps[i] = np.log(np.mean(refdisps)) - np.log(disp)
    return gaps

data, y = make_blobs(n_samples=300, centers=5, random_state=42)
gaps = gap_statistic(data)

plt.plot(range(1, 10), gaps, '-o')
plt.ylabel("Gap Statistic")
plt.xlabel("Number of clusters")
plt.show()

```

# Иерархическая кластеризация
— это метод кластеризации, который строит иерархию кластеров, объединяя или разделяя их последовательно.

Есть два основных типа иерархической кластеризации: агломеративный (объединяющий) и дивизивный (разделяющий).

## Агломеративная иерархическая кластеризация:

Инициализация: Каждый объект представляет собой отдельный кластер, то есть если у нас есть $N$ объектов, то у нас $N$ кластеров.
Объединение: На каждом шаге объединяются два наиболее близких кластера.
Завершение: Процесс продолжается до тех пор, пока все объекты не будут объединены в один кластер.
## Дивизивная иерархическая кластеризация:

Инициализация: Все объекты находятся в одном кластере.
Разделение: На каждом шаге кластер делится на два подкластера.
Завершение: Процесс продолжается до тех пор, пока каждый объект не станет отдельным кластером.
Мера расстояния между кластерами:
Определение "близости" между кластерами является ключевым для иерархической кластеризации. Несколько популярных методов:

Одиночная связь (минимум):
$d(A, B) = \min_{a \in A, b \in B} d(a, b)$, где $d(a, b)$ — расстояние между объектами $a$ и $b$.

Полная связь (максимум):
$d(A, B) = \max_{a \in A, b \in B} d(a, b)$.

Среднее:
$d(A, B) = \frac{1}{|A| \times |B|} \sum_{a \in A} \sum_{b \in B} d(a, b)$.

Центроид:
$d(A, B) = d(\text{centroid}(A), \text{centroid}(B))$, где
 $centroid(A)$
centroid(A) — это центроид (среднее) кластера $A$.

Дендрограмма:
В результате иерархической кластеризации можно построить дендрограмму, которая показывает порядок объединения кластеров. На горизонтальной оси дендрограммы отложены объекты или кластеры, а на вертикальной оси — расстояние (или его производное), на котором произошло объединение.

Определение числа кластеров:
Одним из преимуществ иерархической кластеризации является возможность выбора числа кластеров после выполнения кластеризации, анализируя дендрограмму.

Для определения оптимального числа кластеров можно использовать различные критерии, такие как коэффициент силуэта или внутренние метрики качества кластеризации


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Создаем набор данных
X, _ = make_blobs(n_samples=100, centers=3, random_state=42)

# Выполняем агломеративную иерархическую кластеризацию
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model = model.fit(X)

# Функция для визуализации дендрограммы (используя scipy)
def plot_dendrogram(model, **kwargs):
    # Создаем связки для дендрограммы
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_, counts]).astype(float)
    dendrogram(linkage_matrix, **kwargs)

# Визуализация дендрограммы
plt.title('Dendrogram')
plot_dendrogram(model, truncate_mode='level', p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()

# Если вы хотите визуализировать сами кластеры, вы можете также выполнить:
model = AgglomerativeClustering(n_clusters=3)
model = model.fit(X)
plt.scatter(X[:, 0], X[:, 1], c=model.labels_, cmap='rainbow')
plt.title('Clusters')
plt.show()
```

Сначала мы создаем искусственный набор данных с помощью make_blobs.
Затем мы выполняем агломеративную иерархическую кластеризацию с использованием AgglomerativeClustering.
Мы визуализируем дендрограмму с использованием dendrogram из библиотеки scipy.
После этого мы визуализируем получившиеся кластеры на графике.
Этот код дает представление о том, как можно использовать агломеративную иерархическую кластеризацию и визуализировать результаты.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage

# Создаем набор данных
X, _ = make_blobs(n_samples=100, centers=5, random_state=42)

# Выполняем дивизивную иерархическую кластеризацию с использованием метода "centroid"
Z = linkage(X, method='centroid', optimal_ordering=True)

# Визуализация дендрограммы
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title('Dendrogram for Divisive Hierarchical Clustering')
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```

В этом коде:

Мы создаем набор данных с помощью make_blobs.
Мы используем функцию linkage из scipy для выполнения дивизивной иерархической кластеризации. Метод centroid здесь применяется для определения расстояния между кластерами.
Мы визуализируем результаты с помощью дендрограммы.
Обратите внимание, что функция linkage может использовать различные методы для определения расстояний между кластерами (например, single, complete, average, ward и другие). Выбор метода зависит от конкретной задачи.







- Иерархическая кластеризация включает в себя различные методы, такие как агломеративная (снизу вверх) и дивизивная (сверху вниз). Давайте рассмотрим достоинства и недостатки этих методов.

- Агломеративная иерархическая кластеризация:
Достоинства:

Интерпретируемость: С помощью дендрограммы легко интерпретировать результаты и выбирать оптимальное количество кластеров.
Не требует задания количества кластеров заранее: В отличие от метода k-средних, здесь нет необходимости задавать число кластеров на начальном этапе.
Гибкость: Различные способы определения расстояния (single, complete, average и др.) позволяют учитывать особенности данных.
Недостатки:

Вычислительная сложность: Для больших наборов данных метод может быть очень медленным.
Непереносимость: После того как кластеры объединены, их нельзя разделить на последующих этапах.
Чувствительность к выбору метрики: Результаты могут сильно изменяться в зависимости от выбранной метрики расстояния.
- Дивизивная иерархическая кластеризация:
Достоинства:

Полное деление: В начале работы алгоритма все объекты уже находятся в одном кластере, что может обеспечивать лучшее начальное представление структуры данных.
Гибкость: Как и в агломеративном методе, различные способы определения расстояния позволяют учитывать особенности данных.
Недостатки:

Высокая вычислительная сложность: Этот метод обычно еще менее эффективен, чем агломеративный, особенно на больших наборах данных.
Непереносимость: Аналогично агломеративному методу, разделенные кластеры не могут быть объединены на последующих этапах.
Чувствительность к выбору метрики: Результаты также зависят от выбранной метрики расстояния.
В обоих методах дендрограмма позволяет визуально представить иерархию и делает их очень привлекательными для интерпретируемого машинного обучения. Но вычислительная сложность и некоторые другие недостатки делают их менее популярными для больших наборов данных в сравнении с другими методами кластеризации, такими как k-средних или DBSCAN.

# DBSCAN
Теоретическое введение

Общее описание:
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) – это метод кластеризации, основанный на плотности. В отличие от методов, таких как k-средних, DBSCAN не требует заранее определения числа кластеров. Вместо этого алгоритм формирует кластеры, объединяя плотно расположенные точки и исключая точки, которые далеко от любых других (они считаются шумом).

## Основные понятия:

- Ядро (Core Point):

Точка является ядром, если в её "окрестности" (в радиусе, равном заданному параметру $\epsilon$) содержится не меньше заданного количества точек (параметр $MinPts$, включая саму точку).
Суть в том, что вокруг "ядерной" точки находится достаточно много других точек, что позволяет считать эту область плотно населенной.
- Граничные точки (Border Point):

Если точка не является "ядерной", но находится внутри $\epsilon$-окрестности хотя бы одной "ядерной" точки, то она считается граничной точкой.
Эти точки находятся на границе кластера, но не имеют достаточного числа соседей в своей окрестности, чтобы считаться ядром кластера.
Шумовые точки (Noise Point):

- Точки, которые не являются ни "ядерными", ни "граничными", считаются шумовыми.
Эти точки не принадлежат ни одному кластеру.
Процесс работы алгоритма:

Выберите произвольную точку, которая ещё не была посещена. Определите её соседей на основе $\epsilon$.
Если у этой точки меньше соседей, чем $MinPts$, пометьте её как шумовую точку.
В противном случае начните формирование кластера вокруг этой точки. Добавьте все достижимые "ядерные" точки к кластеру. Граничные точки добавляются к кластеру, но не расширяют его.
Перейдите к следующей необработанной точке.
Преимущества DBSCAN:

Способность находить произвольно формированные кластеры.
Не требует заранее определенного числа кластеров.
Может отфильтровывать шумовые точки.
Недостатки:

Не всегда хорошо справляется с кластерами разной плотности.
В зависимости от выбора параметров $\epsilon$ и $MinPts$, результаты могут сильно отличаться.
DBSCAN - один из самых популярных алгоритмов кластеризации на основе плотности, который хорошо подходит для данных с кластерами сложной формы или для данных с большим количеством шума.


### DBSCAN: Практический пример с использованием Python

Для этого примера мы будем использовать набор данных make_moons из библиотеки sklearn.datasets для иллюстрации того, как DBSCAN может кластеризовать данные сложной формы.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X, _ = make_moons(n_samples=500, noise=0.05, random_state=42)

plt.scatter(X[:, 0], X[:, 1])
plt.title("Generated Data")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis")
plt.title("Clusters with DBSCAN")
plt.show()
```

# Кластеризация с DBSCAN





```python
# Импорт необходимых библиотек и модулей
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN, KMeans

# 1. Изучение данных
# Загрузка набора данных
X, y = make_circles(n_samples=1000, factor=0.3, noise=0.1)
# Визуализация данных
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', edgecolor='black', s=50)
plt.show()
```

```python
# 2. Кластеризация с DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", marker='o', edgecolor='black', s=50)
plt.title('DBSCAN with eps=0.5, min_samples=5')
plt.show()
```

```python
# 4. Сравнение с KMeans
kmeans = KMeans(n_clusters=2)
clusters = kmeans.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", marker='o', edgecolor='black', s=50)
plt.title('KMeans Clustering')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Создание набора данных
data, _ = make_blobs(n_samples=300, centers=2, cluster_std=0.60, random_state=0)
# Добавление аномалий
data[-5:] = np.array([[2, 2], [2, 2.5], [2, 3], [2.5, 2], [2.5, 2.5]])

# Применение DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(data)

# Визуализация данных
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', marker='o', edgecolor='k', s=40)
plt.title("DBSCAN Clustering")
plt.show()
```
