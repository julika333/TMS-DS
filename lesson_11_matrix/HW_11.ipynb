```python
import numpy as np
import matplotlib.pyplot as plt

def convolve2d_manual(image, kernel):
    kernel_height, kernel_width = kernel.shape
    image_height, image_width, num_channels = image.shape

    output = np.zeros((image_height - kernel_height + 1, image_width - kernel_width + 1, num_channels))

    for channel in range(num_channels):
        for y in range(image_height - kernel_height + 1):
            for x in range(image_width - kernel_width + 1):
                output[y, x, channel] = np.sum(
                    image[y:y + kernel_height, x:x + kernel_width, channel] * kernel
                ).clip(0, 255)

    return output.astype(np.uint8)


# Example of a 3x3 convolution kernel
kernel = np.array([[-1, -1, -1],
                   [-1, 10, -1],
                   [-1, -1, -1]])


image = plt.imread('/content/sample_data/Image_4_medium.jpg')
print(image.shape)
print(image)

convolved_image = convolve2d_manual(image, kernel)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(convolved_image)
plt.title("Convolved Image")

plt.show()
```
