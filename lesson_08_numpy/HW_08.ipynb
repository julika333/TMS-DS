```python
import matplotlib.pyplot as plt
img = plt.imread(f"/home/img_fruits/Image_10.jpg")
plt.imshow(img)
```

```python
img.shape
```

```python
data = img.copy()
print(data)
```

```python
def reverse_num(n):
  return 255 - n

reverse_img = reverse_num(data)
print(reverse_img)
```

```python
num_img = 13

fig, ax = plt.subplots(num_img, 2, figsize=(8, num_img * 3))
ax[0,0].set_title("Original img")
ax[0,1].set_title("Reveresed img")

for i in range(num_img):
  img = plt.imread(f"/home/img_fruits/Image_{i+1}.jpg")
  reverse_img = reverse_num(img)
  ax[i,0].imshow(img)
  ax[i,0].axis('off')

  ax[i,1].imshow(reverse_img)
  ax[i,1].axis('off')


plt.show()
```
